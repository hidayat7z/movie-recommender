```python
# VECTORIZATION

from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features=5000,stop_words='english')

#stop_words= is,the,at,on,in,...
```

```python
vectors= cv.fit_transform(new_df['tags'])
```

```python
vectors
```

```python
vectors.toarray()
```

```python
vectors= vectors.toarray()
```

```python
vectors.shape

#no of movies = 4776
#words=5000
```

```python
vectors[0]

# first movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity

## cosine similarity= angular distance b/w two vectors
```

```python
cosine_similarity(vectors)

## cosine similarity score of each with every other movie. SHAPE= 4776x4776
##cos(0)=1
```

```python
simscores= cosine_similarity(vectors)
```

```python
def recommend(movie):
    
    index_pos= new_df[new_df['title']==movie].index[0]
    movie_list= sorted(list(enumerate(simscores[index_pos])),key= lambda x:x[1],reverse=True)[1:6]
    
    for item in movie_list:
        print(new_df.iloc[item[0]]['title'])
```

```python
recommend('Inception')
```

```python
recommend('The Shawshank Redemption')
```

```python
## Printing every movie with simscore
```

```python
def recommend_with_score(movie):
    
    index_pos= new_df[new_df['title']==movie].index[0]
    movie_list= sorted(list(enumerate(simscores[index_pos])),key= lambda x:x[1],reverse=True)[1:6]
    
    for item in movie_list:
        print(new_df.iloc[item[0]]['title'],"-",item[1]*100)
        #print(item[1]*100)
```

```python
recommend_with_score('The Shining')
```

```python
recommend_with_score('The Shawshank Redemption')
```

```python
recommend_with_score('John Carter')
```

```python
recommend_with_score('The Dark Knight Rises')
```
